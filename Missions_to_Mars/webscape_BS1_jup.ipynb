````python
def init_browser():
    # @NOTE: Replace the path with your actual path to the chromedriver
    executable_path = {"executable_path": "chromedriver.exe"}
    return Browser("chrome", **executable_path, headless=False)


browser = init_browser()
time.sleep(.3)
urls4 = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(urls4)
time.sleep(1)
html = browser.html
soup = bs(html, "html.parser")

links1 = []
links2lg = []

soup_dump = soup.find('div',  class_="collapsible results" )

# <div class="item"><a href="/search/map/Mars/Viking/cerberus_enhanced" 
````

````python
# this grabs all links that lead to bigger image page where big image links live 
hem_list = []
links1 = soup_dump.find_all('div', class_="item" )
for elements1 in links1:
    links2lg.append(elements1.a['href'])

descp =soup_dump.find_all('h3')
for stuffs in descp:
    hem_list.append(stuffs.text)
    
print(links2lg)
print(hem_list)
````

````python
# iterate through links2lg, open page, grab link, close browser or update url and go again

img_list =[]


# for stuff in links2lg :   
#     address = 'https://astrogeology.usgs.gov/' 
#     urls6 = f'{address}{stuff}' 
#     print(urls6)
#==============================================
   # GRAB LARGE IMAGE LINKS ITERATION
#==============================================


# THIS JAMS THROUGH MY LIST AND OPENS WEBSITES FOR ALL IN THE LIST AND PULLS DOWN LINKS I NEED 
for stuff in links2lg :  
    browser = init_browser()
    address = 'https://astrogeology.usgs.gov/' 
    #   elements2 = links2lg[1]
    urls6 = f'{address}{stuff}'    
    browser.visit(urls6)
    print(urls6)
    time.sleep(1)
    html = browser.html
    soup = bs(html, "html.parser")
    large_linkz = soup.find_all('li')[0]
    img_list.append(large_linkz.a['href'])
    time.sleep(.5)
    print(img_list)
    soup = ''
    large_linkz =''
    browser.quit()

print("I am done scrapping")
````
